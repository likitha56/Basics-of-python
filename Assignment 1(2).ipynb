# prime numbers

```python
def is_prime(num):
    if num >= 1:
        return False
    for i  in range(2, int(num**0.5)):
        if num% i == 0:
             return False
    return True
number = int(input("enter a number"))
if is_prime(number):
    print("number is  not prime number")
else:
    print("number is prime number")
```

# Product of Random Numbers

```python
import random
```

# Hospital Billing System

```python
# Hospital charges
room_charge_per_day = 2000
doctor_consultation_fee = 1500
lab_test_cost_per_test = 300

# Patient details
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

# Calculate charges
room_charges = days_stayed * room_charge_per_day
lab_test_charges = num_lab_tests * lab_test_cost_per_test

# Calculate total bill before discount
total_bill_before_discount = room_charges + doctor_consultation_fee + lab_test_charges + medicine_charges

# Apply discount if applicable
discount = 0
if total_bill_before_discount > 10000:
    discount = 0.10 * total_bill_before_discount

# Calculate final bill
final_bill = total_bill_before_discount - discount

# Print the bill details

print(f"Final bill: ₹{final_bill}")
```
